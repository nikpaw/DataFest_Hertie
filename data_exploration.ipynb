```python

def add_n_employer(df, column_name, list_range):
    dict_range = {'0-5': 3, '6-24': 15, '25-50': 38, '51-500': 276, '501-5000': 2751, '5001-50000': 27501, '6-50': 28} 

    l = []
    for i in list_range:
        # print(i)
        # if i:
        #     value = dict_range[j]
        # else:
        #     value = 0
        for j in i:
            value = 0
            if not j:
                l.append(0)
            else:
            #chek if the list is empty
                value = dict_range[j]
                break
        l.append(value)
    df[column_name] = l
    return df

list_range = list(map(lambda x: re.findall("(?:=>\"(\d+-\d+)\")",str(x)), list(combined_df.mitarbeiter_staffel)))

add_n_employer(combined_df, 'n_employer', list_range)
#combined_df.to_csv('data/combined_sorted_2_employer.csv')
```
