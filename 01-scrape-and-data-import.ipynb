# Loop on file and cleaning

```python
import os
import pandas as pd
import re

def constr(x):
    try:
        return bool(re.search('[A-Z] [A-Z]', x))
    except:
        return False

directory = 'data/excel'
list_final_df = []
# print(len(os.listdir(directory)))
for i,file in enumerate(os.listdir(directory)):
    filename = os.fsdecode(file)
    o = pd.read_excel('data/excel/' + filename, sheet_name='2_1_1')
    column0 = o.iloc[1][0].split(',')[0].lower()
    list_columns = list(o.iloc[4][1:])
    a = list(map(lambda x: constr(x), list(o['Unnamed: 0'])))
    o = o[a]
    list_columns = [column0] + list_columns
    o.columns = list_columns
    # print(list_columns)
    o['municipality'] = column0 
    o.set_index(column0, inplace=True)
    list_intermidiate_df = []
    for index in  o.index:
        for column in o.columns[:-1]:
            d = {'municipality': [o['municipality'][index]], 'date': [column], 'sector': [index], 'N_employee': [o[column][index]]}
            df = pd.DataFrame(data=d)
            list_intermidiate_df.append(df)
    intermidiate_df = pd.concat(list_intermidiate_df)
    list_final_df.append(intermidiate_df)
    print(f'Precessed {i} files, {len(os.listdir(directory)) - i} files left')


final_df = pd.concat(list_final_df)
```

```python
final_df_badenwurttemberg = final_df
final_df_badenwurttemberg.to_csv('data/final_df_badenwurttemberg.csv')
```
